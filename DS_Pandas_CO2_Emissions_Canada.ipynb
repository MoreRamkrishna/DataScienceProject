```python
# it is follows a_z order
list(df.groupby('Fuel Type'))
```

```python
# fetch AUDI Data
df.groupby('Make').get_group('AUDI')
```

```python
df.groupby('Fuel Type').get_group('Z')
```

```python
df.groupby('Cylinders').get_group(5)
```

```python
# fetch 1 st record of each category
df.groupby('Fuel Type').first()
```

```python
# fetch 2 nd entry
for cat,data in df.groupby('Fuel Type'):
  print(cat,'\n',data[2:3])
  print('\n\n')
```

```python
# can we combine 2 categories
list(df.groupby(['Make','Vehicle Class']))
```

```python
df.groupby(['Make','Vehicle Class']).get_group(('AUDI','SUV - SMALL'))
```

```python
# avg. CO2 Emmision
df.groupby(['Make','Vehicle Class']).get_group(('AUDI','SUV - SMALL'))['CO2 Emissions(g/km)'].mean()
```

```python
df.groupby(['Make','Cylinders'])['CO2 Emissions(g/km)'].mean()
```

```python
# average engine size w.r.t. make
# Agg is aggregate operation

avg = df.groupby('Make')[['Engine Size(L)','CO2 Emissions(g/km)']].agg(np.mean)
avg
```

```python
# color map
avg.style.background_gradient(cmap='Greens')
```
