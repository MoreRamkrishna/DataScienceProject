```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/hakimyameen/Datasets/refs/heads/main/fifa_eda.csv')
df.head(2)
```

```python
df.isna().sum()
```

```python
# drop missing entry
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
# contract valid untile vs value
plt.scatter(df['Contract Valid Until'], df['Value'])
plt.xticks(rotation=90)
plt.show()
```

```python
# histogram:
# univariate plot
# it show rangewise data
# we can use range or bins to set different limits
plt.hist(df['Age'],bins=10)
```

```python
plt.hist(df['Age'],bins=4)
```

```python
# we can specify our own values in bins
plt.hist(df['Age'],bins = [20,30,40,50])
```

```python
# we can also change range
plt.hist(df['Age'], range=[20,40])
```
